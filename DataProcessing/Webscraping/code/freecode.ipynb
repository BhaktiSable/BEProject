```python
freecode=pd.read_csv('classcentral_freecodecamp.csv')
```

```python
freecode.head()
```

```python
freecode.columns=['img url','URL','Title','Description','Duration','Time','Price']
```

```python
freecode.head()
```

```python
urllist=freecode['URL']
```

```python
new_url=[]
for url in urllist:
    domain='https://www.classcentral.com'
    wd.get(url)
    time.sleep(2)
    bs = BeautifulSoup(wd.page_source, 'html.parser')
    desc=bs.find('a',id='btnProviderCoursePage')
    get_url=''
    if desc is not None:
        trim=desc['href']
        final=domain+trim
        print(final)
        wd.get(final)
        time.sleep(9)
        #print(wd.page_source)
        bs = BeautifulSoup(wd.page_source, 'html.parser')
        get_url = wd.current_url
    else:
        desc=bs.find('a',class_='icon-link-charcoal')
        if desc is not None:
            trim=desc['href']
            final=domain+trim
            print(final)
            wd.get(final)
            time.sleep(6)
            #print(wd.page_source)
            bs = BeautifulSoup(wd.page_source, 'html.parser')
            get_url = wd.current_url
        
    # Printing the URL
    print(get_url)
    new_url.append(get_url)
```

```python
freecode['FinalURL']=new_url
```

```python
freecode['Provider']='freeCodeCamp'
```

```python
ratings=[]
for url in urllist:

    get_ratings=''
    url=(url.replace('/classroom/','/course/'))
    #print(url)
    wd.get(url)
    time.sleep(2)
    bs = BeautifulSoup(wd.page_source, 'html.parser')
    div1=bs.find('div',id='reviews-contents')
    if div1 is not None:
        
        div2=div1.find('div',class_='margin-top-small')
        if div2 is not None :
            div2=div2.find('strong',class_='weight-bold')
            if div2 is not None :
                get_ratings=div2.text
    print(get_ratings)
    ratings.append(get_ratings)
```

```python
freecode['Ratings']=ratings
```
