```python
from bs4 import BeautifulSoup
import requests
import time
import re

import pandas as pd
import numpy as np

# set options to be headless, ..
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException   
```

```python
# Initialize chrome driver
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')

wd = webdriver.Chrome('chromedriver',options=options)
wd.implicitly_wait(30)
```

```python
df=pd.read_csv(r'C:\Users\nehab\Desktop\Python\final datasets\LinkedIn_data.csv')
```

```python
df.info()
```

```python
level=[]
urllist=df['FinalUrl']
df['Level']=''
```

```python
i=0
for url in urllist:
    wd.get(url)
    time.sleep(1)
    diff=''
    bs = BeautifulSoup(wd.page_source, 'html.parser')
    div=bs.find('h2',class_='top-card-layout__headline break-words font-sans text-md leading-open text-color-text')
    if div is not None:
        div2=div.find('span',text=re.compile('Skill level'))
    #print(div2)
    if div2 :
        diff=div2.text.replace('Skill level:','').strip()
        print(diff)
    level.append(diff)
    df['Level'][i]=diff
    i+=1
```

```python
df.head()
```

```python
df.to_csv('LinkedIn_data____1.csv',index=False,encoding='utf_8_sig')
```

```python
df['Numenroll']=''
df['RRRR']=''
skills=[]
df['skill learn']=''
```

```python
j=0
ratings=[]
enrollnum=[]
for url in urllist:
    wd.get(url)
    time.sleep(1)
    rat=''
    bs = BeautifulSoup(wd.page_source, 'html.parser')
    div=bs.find('h3',class_='ratings-summary__header')
    if div is not None:
        div2=div.find('span',class_='ratings-summary__rating-average')
        rat=(div2.text)
    
    enroll=''
    
    div=bs.find('div',class_='top-card__headline-row')
    if div is not None:
        div2=div.find('span',text=re.compile('Liked by '))
    #print(div2)
        if div2 :
            enroll=div2.text.replace('Liked by ','').strip()
            print(enroll)
            
    skill=''
   
    div=bs.find('ul',class_='course-skills__skill-list')
    #print(div)
    if div is not None:
        for i in div:
            tt=i.text
            tt=(tt.strip())
            if tt !='':
                skill=(tt)+','+skill
    skills.append(skill)
    
    df['skill learn'][j]=skill
    
    print(skill)
    
    enrollnum.append(enroll)
    df['Numenroll'][j]=enroll
    
    ratings.append(rat)
    df['RRRR'][j]=rat
    
    j+=1
```

```python
df.to_csv('LinkedIn_data____1.csv',index=False,encoding='utf_8_sig')
```
