```python
courses_difficulty_levels=[]
for url in new_url:
    wd.get(url)
    bs = BeautifulSoup(wd.page_source, 'html.parser')
    level=bs.find('div',class_='InformationWidget-widgetWrapper_6ZMLn')
    if level is not None:
        for x in level.find_all('li',class_='listItemWithIcon-module_wrapper__q2laj listItemWithIcon-module_medium__2k3BT'):
            if x is not None:
                level=x.text
    else:
        level=''
    print(level)
    courses_difficulty_levels.append(level)
```
